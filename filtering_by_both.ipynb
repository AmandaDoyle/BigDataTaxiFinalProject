```python
import pandas as pd
pd.set_option('display.mpl_style', 'default')
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
dfe = pd.read_json("EveningDropoffs.json", orient="index")
```

```python
dfe[2010.0]=(dfe[2010.0]-dfe[2010.0].mean())/dfe[2010.0].std()
dfe[2011.0]=(dfe[2011.0]-dfe[2011.0].mean())/dfe[2011.0].std()
dfe[2012.0]=(dfe[2012.0]-dfe[2012.0].mean())/dfe[2012.0].std()
dfe[2013.0]=(dfe[2013.0]-dfe[2013.0].mean())/dfe[2013.0].std()
```

```python
dfe.dropna(inplace=True)
dfclean= dfe[[2010.0, 2012.0]]
```

```python

deltas = dfclean.pct_change(axis = 1)
deltas = deltas.drop(2010.0, axis = 1)
deltas.dropna()
deltas.describe(percentiles=[0.05, 0.1, 0.25, 0.5, 0.75, 0.85, 0.9, 0.95])
```

```python
high_change = deltas[deltas>=deltas.quantile(0.95)]
high_change = high_change.dropna(how='all')
high_change
```

```python
intersections = pd.read_csv('intersections.csv', header = None)
```

```python
intersections.index.name = 'int_id'

high_change.index.name = 'int_id'
```

```python
with_int = high_change.join(intersections)
with_int.to_csv('deltaswithintnorm20102012.csv')
with_int.head()
```
